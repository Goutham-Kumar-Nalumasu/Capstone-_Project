# Training the Linear regression model
* let train your regression model we will need to first split the data into x array that contains the features to train on and y array with the target variable

```python
# creating the x training variable 
x=data1[['Age','Base_pay','Unit_Sales','Total_Sales']]
x
```

```python
# creating the y target variable is  the that your final aganda to predict the salary
#  by using ml algorithms comparing the actual salary vs predicted salary
y=data1['Salary']
y.head()
```

# Train test split
*  let split  the data into a training set and a testing set we will train out model on the training set and then use the test set to evalualte the model

```python
# importing  train_test_split libary 
from sklearn.model_selection import train_test_split
```

```python
# spliting the model into Training  and test set 
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
# shape of x_train and x_test
print('number of training samples',X_train.shape[0])
print('number of test sapmles',X_test.shape[0])
```

# Creating and Training the model

```python
# importing the linear regression libary
from sklearn.linear_model import LinearRegression

# we instante a instants for linear regression object
lr=LinearRegression()
```

```python
# x and y are fitted into the model
lr.fit(X_train,y_train)
```

```python
# predict function is used to create array of target values    

pred=lr.predict(X_train)
pred.shape  # number of  rows
```

# Rsquared and MSE 
## RMSE:

* RMSE (Root Mean Squared Error) represents square root of  difference between the original and predicted values extracted by squared the  average difference over the data set.




**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$






## Rsquared

* Rsquares is the proportion of the variance  in the responce variable that can be explanied by the predictor variables in a linear refression model  the value of rsquared is range from -1 to 1





* r2 =1-(sum square regression error/ sum square total error) 


**The lower mse and higher Rsquared values are desirable**

```python
# Evaluate the RMSE
from sklearn.metrics import mean_squared_error             # importing MSE libary
from sklearn.metrics import r2_score                           #importing r2 square libaru
rm_s_e=(np.sqrt(mean_squared_error(y_train,pred)))   # we instante a instants for linear regression object
print('Root mean squared error is:',rm_s_e)
# Evaluating with the R2 which is a metrix model for model prediation accuarcy   
print('The R-squared is:',r2_score(y_train,pred))
```

```python
# Evaluate with MSE 
# cross validation score is useful in assessing the effectiveness of the model
from sklearn.model_selection import cross_val_score

# returns the predicted value of each input
from sklearn.model_selection import cross_val_predict
Rcross= cross_val_score(lr,X_train,y_train,cv=5)
print('The Rcross validation accuarcy  is :',(Rcross.mean(),Rcross.std()))
```
